```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Problem 1

```python
def prob1():
    """
    Create 3 visualizations of the crime data set.
    One of the visualizations should be a histogram.
    The visualizations should be clearly labelled and easy to understand.
    """
    
    # Read in data
    crime_data = pd.read_csv("crime_data.csv", index_col = 0)
    
    # First Plot
    crime_data.plot(y = "Total", title = "Total Crime vs. Time")
    
    # Second Plot
    crime_data[["Violent", "Property"]].hist(grid= False, bins = 10)

    # Third Plot
    crime_data.plot(y= ['Murder', 'Forcible Rape', 'Robbery', 'Aggravated Assault'], subplots = True, layout = (2,2), sharey = True, style = ['-', '-', '-', '-'], title = "Different Plots")
    

    return crime_data
   

    
```

```python
prob1()
```

# Problem 2

```python
def prob2():
    """
    Use crime_data.csv to plot the trends between Larceny and
        1. Violent
        2. Burglary
        3. Aggravated Assault
    Each graph should be clearly labelled and readable.
    One of these variables does not have a linear trend with Larceny.
    Return a string identifying this variable.
    """
    
    # Read in data
    crime_data = pd.read_csv("crime_data.csv", index_col = 0)
    
    # Plot trends
    crime_data.plot(kind = "scatter", x = "Larceny", y = "Violent", alpha = 0.8, xlim = (0, max(crime_data['Larceny']) + 1), ylim = (0, max(crime_data['Violent']) + 1), title = "Larceny vs. Violent Crime") 
    crime_data.plot(kind = "scatter", x = "Larceny", y = "Burglary", alpha = 0.8, xlim = (0, max(crime_data['Larceny']) + 1), ylim = (0, max(crime_data['Burglary']) + 1), title = "Larceny vs. Burglary")
    crime_data.plot(kind = "scatter", x = "Larceny", y = "Aggravated Assault", alpha = 0.8, xlim = (0, max(crime_data['Larceny']) + 1), ylim = (0, max(crime_data['Aggravated Assault']) + 1), title = "Larceny vs. Aggravated Assault")

    # Return variable that does not have linear trend
    No_Linear_Trend = "Burglary"
    return No_Linear_Trend
   
```

```python
prob2()
```

# Problem 3

```python
def prob3():
    """
    Use crime_data.csv to display the following distributions.
        1. The distributions of Burglary, Violent, and Vehicle Theft 
            as box plots
        2. The distributions of Vehicle Thefts against Robberies as
            a hexbin plot.
    All plots should be labelled and easy to read.
    """
    
     # Read in data
    crime_data = pd.read_csv("crime_data.csv", index_col = 0)

    # Plot box plot
    crime_data.plot(kind = "box", y = ["Burglary", "Violent", "Vehicle Theft"], ylabel = "Number of Incidents")

    # Plot Hexbin
    crime_data.plot(kind = "hexbin", x = "Vehicle Theft", y = "Robbery", gridsize = 20, xlabel = "Vehicle Thefts", ylabel = "Robberies", sharex= False)
    
    
```

```python
prob3()
```

```python
def prob4():
    """
    Create 3 plots that compare variables or universities.
    These plots should answer questions about the data, 
    (e.g. What is the distribution of graduation rates? Do schools 
    with lower student to faculty ratios have higher tuition costs? 
    etc.). These plots should be easy to understand and have clear 
    titles, variable names, and citations.
    """
    
    # Read in data
    college_data = pd.read_csv("college.csv", index_col=0)

    # Data source citation
    data_source_citation = "Citation:  https://cran.r-project.org/web/packages/ISLR/ISLR.pdf"

    # Plot #1 - Professor PhD Rate vs. Graduation Rate
    plt.figure(figsize=(8, 6))
    plt.scatter(college_data["PhD"], college_data["Grad.Rate"], alpha=0.8)
    plt.title("Professor PhD Rate vs. Graduation Rate")
    plt.xlabel("PhD Rate")
    plt.ylabel("Graduation Rate")
    plt.text(0.5, -0.2, data_source_citation, ha='center', transform=plt.gca().transAxes)
    plt.show()

    # Plot #2 - Boxplot of Alumni Donation Rates
    plt.figure(figsize=(8, 6))
    college_data.boxplot(column=["perc.alumni"])
    plt.title("Distribution of Alumni Donation Rates at Universities")
    plt.ylabel("Percentage of Alumni that Donated")
    plt.text(0.5, -0.2, data_source_citation, ha='center', transform=plt.gca().transAxes)
    plt.show()

    # Plot #3 - Hexbin Plot of Personal Spending vs. Book Spending
    plt.figure(figsize=(8, 6))
    plt.hexbin(college_data["Personal"], college_data["Books"], gridsize=20, cmap='Blues')
    plt.title("Personal Spending vs. Book Spending")
    plt.xlabel("Personal Spending")
    plt.ylabel("Book Spending")
    plt.xlim(min(college_data["Personal"]) - 1, 3000)
    plt.ylim(0, 1000)
    plt.text(0.5, -0.2, data_source_citation, ha='center', transform=plt.gca().transAxes)
    plt.show()

    
```

```python
prob4()
```
