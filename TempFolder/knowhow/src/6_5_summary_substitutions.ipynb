```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

files = [  "../results/substitutions/bodyfat_substitution_results.csv",
  "../results/substitutions/trig_substitution_results.csv",
  "../results/substitutions/hdl_substitution_results.csv",
  "../results/substitutions/ldl_substitution_results.csv",
  "../results/substitutions/hba1c_substitution_results.csv",
  "../results/substitutions/bmi_substitution_results.csv",
  "../results/substitutions/waistcirumference_substitution_results.csv"]

outcome_map = {'bodyfat':'Body Fat (%)',
                     'bmi':'BMI (kg/m^2)',
                     'waistcirumference':'Waist Circumference (cm)',
                     'trig':'Triglycerides (mmol/L)',
                     'ldl':'LDL (mmol/L)',
                     'hdl':'HDL (mmol/L)',
                     'hba1c':'HbA1c (%)'
                     }

predictor_map = {'nmes':'Non-Milk Extrinisic Sugars (g)',
                       'intrinsic':'Intrinsic Sugars (g)',
                       'sugar':'Intrinsic Sugars (g)',

                       'starch':'Starch (g)',
                       'fiber':'Fiber (g)'}

def determine_trend_symbol(lower, upper):
    """Determine trend symbol based on data trend"""

    if lower > 0 and upper > 0:
        return '↑'
    elif lower < 0 and upper < 0:
        return '↓'
    else:
        return '⌀'


final_df = pd.DataFrame(columns = ['Outcome','Intake decresed', 'Intake increased', 'result'])
for file in files:
    df = pd.read_csv(file)
    outcome = file.split('/')[3].split('_')[0]
    for col in df:
        row = []
        values = df[col].to_list()
        values.sort()
        l = len(values)
        mean = np.round(np.mean(values),2)
        lower_idx = int(l//100 * 2.5)
        upper_idx = int(l//100 * 97.5)
        lower = np.round(values[lower_idx],2)
        upper = np.round(values[upper_idx],2)
        s = col.split('_')
        predictor = s[1]
        substitute = s[-1]
        trend = determine_trend_symbol(lower, upper)
        row = [outcome_map[outcome], predictor_map[predictor], 
               predictor_map[substitute], trend + '\n(' +str(lower)+'; '+ str(upper)+')']
        final_df.loc[len(final_df)] = row
        
    

    
final_df.to_csv('../results/substitutions/substitution_summary.csv', index=False)
```
