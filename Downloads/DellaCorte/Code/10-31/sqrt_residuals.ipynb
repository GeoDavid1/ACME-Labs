```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Define the path to your Excel file
path = '/Users/user/Downloads/DellaCorte/XSLX/Merged.xlsx'
sheet_name = 'Sqrt_Merged_Averages'

# Load the data from the Excel file into a Pandas DataFrame
df = pd.read_excel(path, sheet_name=sheet_name)

# Find columns that start with 'Cumulative'
cumulative_columns = [col for col in df.columns if col.startswith('Cumulative')]

# Specify the reference column
reference_column = 'Cumulative Average for intake_energy_megajoules'

# Create an empty DataFrame to store residuals
residual_df = pd.DataFrame()

# Add the 'participant_id' column to the left side of the DataFrame
residual_df['participant_id'] = df['participant_id']

# Iterate through each cumulative column
for col in cumulative_columns:
    if col != reference_column:
        # Extract the independent variable (reference column)
        independent_variable = df[[reference_column]]
        
        # Extract the dependent variable (cumulative column)
        dependent_variable = df[col]
        
        # Drop rows with NaN values in both independent and dependent variables
        df_cleaned = pd.concat([independent_variable, dependent_variable], axis=1).dropna()
        independent_variable_cleaned = df_cleaned[reference_column]
        dependent_variable_cleaned = df_cleaned[col]
        
        # Create a linear regression model
        model = LinearRegression()
        
        # Fit the model to predict the dependent variable using the independent variable
        model.fit(independent_variable_cleaned.values.reshape(-1, 1), dependent_variable_cleaned)
        
        # Calculate the residuals
        residuals = dependent_variable_cleaned - model.predict(independent_variable_cleaned.values.reshape(-1, 1))
        
        # Store the residuals in the DataFrame
        residual_df[col + ' - Residual'] = residuals

# Save the DataFrame with residuals to an Excel file
output_path = 'residuals_sklearn_linear_regression.xlsx'
residual_df.to_excel(output_path, index=False)
print(f'Residuals saved to {output_path}')





```
