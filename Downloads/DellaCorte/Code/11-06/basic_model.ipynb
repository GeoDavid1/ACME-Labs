```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

# Define the file path as a variable
file_path = '/Users/user/Downloads/DellaCorte/Sqrt_Final_Averages.xlsx'

def read_data(file_path):
    # Read data from the Excel file into a pandas DataFrame
    df = pd.read_excel(file_path)
    return df

def get_linear_regression(data, X_variable, y_variable):
    # Extract the X and Y data from the DataFrame
    X = data[X_variable]
    Y = data[y_variable]

    # Drop rows with missing values in either X or Y
    df_cleaned = data.dropna(subset=[X_variable, y_variable])

    # Extract the cleaned X and Y data
    X_cleaned = df_cleaned[X_variable]
    Y_cleaned = df_cleaned[y_variable]

    # Add a constant term (intercept) to the independent variable
    X_cleaned = sm.add_constant(X_cleaned)

    # Create and fit the linear regression model using cleaned data
    model = sm.OLS(Y_cleaned, X_cleaned).fit()

    # Return the regression model
    return model, df_cleaned

def plot_regression(model, X_variable, y_variable, data):
    # Create a regression plot
    plt.figure(figsize=(8, 6))
    sns.regplot(x=X_variable, y=y_variable, data=data, ci=None, line_kws={"color": "red"})

    # Label the axes and title
    plt.xlabel(X_variable)
    plt.ylabel(y_variable)
    plt.title(f"Linear Regression Plot\n({X_variable} vs. {y_variable})")

    # Show the plot
    plt.show()

# Read the data from the Excel file
data = read_data(file_path)




    
    
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
from itertools import product

# Read the data from the Excel file
data = read_data(file_path)

# Define X and Y variables
X_variables = ['centre_id', 'age_at_study', 'Sqrt of Cumulative Average for intake_nmes']
Y_variables = ['Sqrt of ecid_hba1c_percent', 'Sqrt of cid_moissl_fat_mass_percent', 'Sqrt of ecid_ldl', 'Sqrt of ecid_hdl']

# Perform linear regression analysis for every combination
for X_variable, Y_variable in product(X_variables, Y_variables):
    # Perform linear regression analysis
    regression_model, cleaned_data = get_linear_regression(data, X_variable, Y_variable)

    # Print a summary of the regression results
    print(f"Regression results for {X_variable} vs {Y_variable}:")
    print(regression_model.summary())

    # Create and display the regression plot with labels and a title
    plot_regression(regression_model, X_variable, Y_variable, cleaned_data)
    plt.show()
```
