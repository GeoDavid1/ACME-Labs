```python
import pandas as pd
import numpy as np

CUMULATIVE_RESIDUALS_PATH = '/Users/user/Downloads/DellaCorte/knowhow/data/cleaned data/Cumulative_Residuals_Averages.xlsx'

# Read the DataFrame from the Excel file
df = pd.read_excel(CUMULATIVE_RESIDUALS_PATH)

# Find rows where 'cid_moissl_recorded_sex' == 0
cid_moissl_0_indices = df[df['cid_moissl_recorded_sex'] == 0].index

# Randomly select 5 indices for sex == 1 and 5 for sex == 2
np.random.seed(42)  # Set a seed for reproducibility
indices_sex_1 = np.random.choice(cid_moissl_0_indices, 5, replace=False)
remaining_indices = set(cid_moissl_0_indices) - set(indices_sex_1)
indices_sex_2 = np.random.choice(list(remaining_indices), 5, replace=False)

# Assign sex == 1 for the selected indices and sex == 2 for the remaining
df.loc[indices_sex_1, 'cid_moissl_recorded_sex'] = 1
df.loc[indices_sex_2, 'cid_moissl_recorded_sex'] = 2

# Save the modified DataFrame to the original file path
df.to_excel(CUMULATIVE_RESIDUALS_PATH, index=False)
```
